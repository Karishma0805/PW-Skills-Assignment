4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
variable across different categories.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a dataset with categories and corresponding numerical values
np.random.seed(42)  # For reproducibility
categories = ['Category A', 'Category B', 'Category C', 'Category D']

# Create numerical data for each category (random data with different means)
data = {
    'Category A': np.random.normal(10, 5, 100),  # Mean=10, std=5
    'Category B': np.random.normal(20, 5, 100),  # Mean=20, std=5
    'Category C': np.random.normal(30, 5, 100),  # Mean=30, std=5
    'Category D': np.random.normal(25, 5, 100)   # Mean=25, std=5
}

# Convert the data into a suitable format for plotting
category_data = []
category_labels = []

for category in categories:
    category_data.extend(data[category])
    category_labels.extend([category] * len(data[category]))

# Plotting the box plot
plt.figure(figsize=(8, 5))
sns.boxplot(x=category_labels, y=category_data, palette="Set2")
plt.title('Distribution of Numerical Values Across Categories')
plt.xlabel('Categories')
plt.ylabel('Numerical Values')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a synthetic dataset with correlated features
np.random.seed(42)  # For reproducibility

# Create random data with correlations
n_samples = 1000
X1 = np.random.normal(0, 1, n_samples)
X2 = 0.8 * X1 + np.random.normal(0, 0.5, n_samples)  # X2 is correlated with X1
X3 = 0.6 * X1 + 0.4 * X2 + np.random.normal(0, 0.5, n_samples)  # X3 is correlated with X1 and X2

# Combine into a dataset
data = np.vstack([X1, X2, X3]).T  # Stack them into columns

# Compute the correlation matrix
corr_matrix = np.corrcoef(data, rowvar=False)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, cbar=True, xticklabels=['X1', 'X2', 'X3'], yticklabels=['X1', 'X2', 'X3'])
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a meshgrid
x, y = np.meshgrid(x, y)

# Calculate z = sin(sqrt(x^2 + y^2))
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Plotting the 3D surface plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot surface
surf = ax.plot_surface(x, y, z, cmap='viridis')

# Add title and labels
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Add color bar for better understanding of the surface height
fig.colorbar(surf)

# Show plot
plt.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.


```python
from bokeh.plotting import figure, show
import numpy as np

# Generate x and y values
x = np.linspace(0, 10, 100)  # x-values from 0 to 10
y = np.sin(x)  # y-values as the sine of x

# Create a Bokeh plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer with the x and y values
p.line(x, y, legend_label="sin(x)", line_width=2, line_color="blue")

# Show the plot
show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd

# Generate random data
np.random.seed(42)  # For reproducibility
n = 100  # Number of points
x = np.random.rand(n) * 10  # Random x values between 0 and 10
y = np.random.rand(n) * 10  # Random y values between 0 and 10
sizes = np.random.randint(5, 50, size=n)  # Random sizes for markers
colors = np.random.rand(n)  # Random color values for the markers

# Create a DataFrame for easier handling
df = pd.DataFrame({
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
})

# Create a Bokeh scatter plot
p = figure(title="Bokeh Scatter Plot", x_axis_label='x', y_axis_label='y')

# Add scatter renderer
p.scatter(x='x', y='y', size='sizes', color='colors', source=df, legend_label="Random Data", fill_alpha=0.6)

# Add title and labels
p.title.text_font_size = '16pt'
p.xaxis.axis_label_text_font_size = '12pt'
p.yaxis.axis_label_text_font_size = '12pt'

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
import numpy as np

# Generate random data
data_hist = np.random.randn(1000)

# Compute histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram of Random Data",
           x_axis_label='Value',
           y_axis_label='Frequency',
           plot_height=400,
           plot_width=600)

# Add bars (quad is used to draw rectangles in Bokeh)
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="skyblue", line_color="white", alpha=0.7)

# Show the plot
show(p)
```
